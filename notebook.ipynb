### **Pertanyaan 1: Kota Apa yang Memiliki Jumlah Pembeli Terbesar?**

Untuk menentukan strategi pemasaran, kita perlu memahami seberapa populer / dikenal produk kita di masyarakat. Salah satu caranya dengan menelaah jumlah pembeli yang tersebar di masing - masing kota. Untuk itu, berikut ditampilkan Top 5 & Bottom 5 Kota berdasarkan Jumlah Pembeli

```python
# Menampilkan dua figur sejajar
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,6))

# Menampilkan 5 Kota dengan Jumlah Pembeli Terbesar di Bagian Kiri Subplot
ax1.bar(top_5_city_key, top_5_city_value, color='green')
ax1.set_xticklabels(top_5_city_key, rotation=45)
ax1.set_xlabel('Nama Kota')
ax1.set_ylabel('Jumlah Pembeli')
ax1.set_title('Top 5 - Kota dengan Jumlah Pembeli Terbesar')

# Menampilkan 5 Kota dengan Jumlah Pembeli Terkecil di Bagian Kanan Subplot
ax2.bar(bottom_5_city_key, bottom_5_city_value, color='red')
ax2.set_xticklabels(bottom_5_city_key, rotation=45)
ax2.set_xlabel('Nama Kota')
ax2.set_ylabel('Jumlah Pembeli')
ax2.set_title('Bottom 5 - Kota dengan Jumlah Pembeli Terkecil')

# Membuat Judul
plt.suptitle("Kota dengan Jumlah Pembeli Terbesar & Terkecil", fontsize=20)

# Menampilkan Visualisasi
plt.show()
```

Dari diagram di atas dapat dilihat bahwa sau paulo menjadi kota dengan jumlah pembeli terbesar, sedangkan rata rata kota terendah memiliki hanya satu pembeli.

### **Pertanyaan 2: Berapa Distribusi Jumlah Pembeli yang Merupakan Repeater dan Non Repeater?**

Salah satu bentuk keberhasilan platform e - commerce adalah membuat pembelinya kembali untuk membeli produk, baik barang yang sama ataupun mencoba membeli barang baru. Untuk mengetahui keberhasilan marketing ini, kita dapat melihat melalui persebaran pembeli Repeater & Non Repeater.

```python
plt.title('Distribusi Jumlah Pembeli')
plt.pie(repeater_value, labels=repeater_key, colors=sns.color_palette('Set3'),
        autopct='%1.1f%%', shadow=True, startangle=180)

plt.axis('equal')
plt.show()
```

Distribusi pembeli yang dikategorikan sebagai Repeater masih tergolong rendah. Strategi marketing perlu diperkuat untuk membuat pembeli mempercayakan pembelian barang kepada e-commerce ini.

### **Pertanyaan 3: Apakah Besaran Biaya Kirim Berkorelasi dengan Total Harga Pembelian Barang?**

Biaya Kirim Merupakan salah Satu hal yang biasa menjadi pertimbangan bagi pembeli untuk membeli baran pada Platform E - Commerce. Melalui analisis explanatory ini kita berusaha memahami korelasi antara biaya kirim dan total harga pembelian barang.

```python
cmap = sns.diverging_palette(10, 220, as_cmap=True)
# plt.matshow(group_filtered_order.corr(), color=cmap)
# plt.show()

sns.heatmap(
    group_filtered_order.corr(),
    cmap=cmap,
    vmax=1,
    vmin=-1,
    center=0,
    square=True,
    linewidths=.5,
    cbar_kws={"shrink": .5}
)
```

Bila dilihat dari diagram korelasi di atas, biaya pengantaran dan total biaya memiliki korelasi positif. Hal ini menunjukkan bahwa kedua variabel sejalan.
